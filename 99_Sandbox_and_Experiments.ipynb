```python
import torch
from config import cfg
from tokenizers import Tokenizer
from dataset import build_dataloaders

print("=" * 60)
print("DATALOADER REAL-DATA INSPECTION")
print("=" * 60)

# 1. Update config dynamically
tokenizer = Tokenizer.from_file("bpe_tokenizer.json")
cfg.vocab_size = tokenizer.get_vocab_size()

# 2. Build the actual DataLoaders using real data files
print("Building DataLoader...")
train_loader, val_loader = build_dataloaders(
    train_path="data/train.jsonl",
    val_path="data/val.jsonl",
    tokenizer_path="bpe_tokenizer.json",
    batch_size=4  # Small batch size
)

print("\nExecuting the 'for' loop...\n")

# 3. Iterate through exactly 2 batches
for batch_idx, (src, tgt_in, tgt_out) in enumerate(train_loader):
    print("=" * 60)
    print(f"BATCH {batch_idx + 1}")
    print("=" * 60)

    print(f"[SHAPES]")
    print(f"  src:     {src.shape}     -> [Batch, MaxSrcLen]")
    print(f"  tgt_in:  {tgt_in.shape}     -> [Batch, MaxTgtLen]")
    print(f"  tgt_out: {tgt_out.shape}     -> [Batch, MaxTgtLen]")

    print("\n[RAW TENSORS - ROW 0]")
    print(f"  src:     {src[0].tolist()}")
    print(f"  tgt_in:  {tgt_in[0].tolist()}")
    print(f"  tgt_out: {tgt_out[0].tolist()}")

    print("\n[DECODED TEXT - ROW 0]")
    # We use tokenizer.decode() to turn the integers back to text.
    # Note: decode() naturally ignores special tokens, so we print the IDs first to see them.
    print(f"  Source (English): {tokenizer.decode(src[0].tolist())}")

    # tgt_in has the SOS token (1) at the front
    print(f"  Target IN  (SOS): ID {tgt_in[0][0].item()} -> {tokenizer.decode(tgt_in[0].tolist())}")

    # tgt_out has the EOS token (2) at the end (before any padding)
    # We strip padding (0) to find the true end of the sentence
    non_padded_length = (tgt_out[0] != cfg.PAD_IDX).sum().item()
    last_token_id = tgt_out[0][non_padded_length - 1].item()
    print(f"  Target OUT (EOS): {tokenizer.decode(tgt_out[0].tolist())} <- ID {last_token_id}")

    print("\n[PADDING CHECK]")
    print(f"  Does src contain padding (0)? {'Yes' if cfg.PAD_IDX in src else 'No'}")

    # Stop after inspecting 2 batches to compare their shapes!
    if batch_idx == 1:
        break
```
